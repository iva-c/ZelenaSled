```python
# Get bounding box for Ljubljana

import osmnx as ox


city_name = "Ljubljana, Slovenia"

# Get the boundary of Ljubljana(GeoDataFrame) from OpenStreetMaps
gdf = ox.geocode_to_gdf(city_name)
min_lat, min_lon, max_lat, max_lon = gdf.total_bounds

print(min_lat, min_lon, max_lat, max_lon ) # 14.408617 45.9740637 14.755332 46.1459968
```

```python
# get graphs from OpenStreetMaps

# Get the graph for cycling
G_bike = ox.graph_from_place(city_name, network_type='bike')
ox.save_graphml(G_bike, filepath="/ljubljana_bike.graphml")

# Get the graph for walking
G_walk = ox.graph_from_place(city_name, network_type='walk')
ox.save_graphml(G_walk, filepath="/ljubljana_walk.graphml")

```

```python
# Plot cycling graph
ox.plot_graph(G_bike, bgcolor='white', node_color='blue', edge_color='green')

# Plot walking graph
ox.plot_graph(G_walk, bgcolor='white', node_color='red', edge_color='orange')
```
